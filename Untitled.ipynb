```python
import pandas as pd
import numpy as np
import os

# ──────────────────────────────────────────────
# STEP 0: Load Data
# ──────────────────────────────────────────────
# Option A: Load from local file
# df = pd.read_excel("online_retail.xlsx")

# Option B: Load from GitHub (replace with your raw URL)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx"
print("Loading data...")
df = pd.read_excel(url)
print(f"Loaded: {df.shape[0]} rows × {df.shape[1]} columns")

# ──────────────────────────────────────────────
# STEP 1: Remove Duplicate Rows
# ──────────────────────────────────────────────
before = len(df)
df.drop_duplicates(inplace=True)
print(f"Step 1 – Duplicates removed: {before - len(df)}")

# ──────────────────────────────────────────────
# STEP 2: Drop Missing Customer IDs
# ──────────────────────────────────────────────
before = len(df)
df.dropna(subset=['CustomerID'], inplace=True)
print(f"Step 2 – Missing CustomerID removed: {before - len(df)}")

# ──────────────────────────────────────────────
# STEP 3: Fill Missing Descriptions
# ──────────────────────────────────────────────
df['Description'] = df['Description'].fillna('Unknown')
print("Step 3 – Missing Description filled with 'Unknown'")

# ──────────────────────────────────────────────
# STEP 4: Remove Cancellation Invoices (start with 'C')
# ──────────────────────────────────────────────
before = len(df)
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
print(f"Step 4 – Cancellation invoices removed: {before - len(df)}")

# ──────────────────────────────────────────────
# STEP 5: Remove Negative / Zero Quantity & Price
# ──────────────────────────────────────────────
before = len(df)
df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]
print(f"Step 5 – Invalid Quantity/Price removed: {before - len(df)}")

# ──────────────────────────────────────────────
# STEP 6: Fix Data Types
# ──────────────────────────────────────────────
df['CustomerID']  = df['CustomerID'].astype(int).astype(str)
df['InvoiceNo']   = df['InvoiceNo'].astype(str).str.strip()
df['StockCode']   = df['StockCode'].astype(str).str.strip().str.upper()
df['Description'] = df['Description'].astype(str).str.strip().str.title()
df['Country']     = df['Country'].astype(str).str.strip()
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate']).dt.strftime('%Y-%m-%d')
print("Step 6 – Data types fixed and whitespace cleaned")

# ──────────────────────────────────────────────
# STEP 7: Rename Columns (clean names)
# ──────────────────────────────────────────────
df.rename(columns={
    'InvoiceNo'  : 'Invoice',
    'UnitPrice'  : 'Price',
    'CustomerID' : 'Customer ID'
}, inplace=True)

# ──────────────────────────────────────────────
# STEP 8: Add TotalAmount Column
# ──────────────────────────────────────────────
df['TotalAmount'] = (df['Quantity'] * df['Price']).round(2)
df['Price']       = df['Price'].round(2)
print("Step 8 – TotalAmount column added")

# ──────────────────────────────────────────────
# STEP 9: Remove Outliers (Quantity > 99.9th percentile)
# ──────────────────────────────────────────────
q999 = df['Quantity'].quantile(0.999)
before = len(df)
df = df[df['Quantity'] <= q999]
print(f"Step 9 – Outliers removed (Quantity > {q999:.0f}): {before - len(df)}")

# ──────────────────────────────────────────────
# STEP 10: Encode Country to reduce file size
# ──────────────────────────────────────────────
country_map = {c: f'C{i:02d}' for i, c in enumerate(df['Country'].unique())}
reverse_map = {v: k for k, v in country_map.items()}
df['Country'] = df['Country'].map(country_map)

# Save country legend
legend_df = pd.DataFrame(list(reverse_map.items()), columns=['Code', 'Country'])
legend_df.to_csv('country_legend.csv', index=False)
print("Step 10 – Countries encoded, legend saved to country_legend.csv")

# ──────────────────────────────────────────────
# STEP 11: Sample to ~20MB & Reset Index
# ──────────────────────────────────────────────
df = df.sample(frac=0.72, random_state=42).sort_index().reset_index(drop=True)
print(f"Step 11 – Sampled to {len(df)} rows")

# ──────────────────────────────────────────────
# STEP 12: Save to CSV
# ──────────────────────────────────────────────
output_path = 'online_retail_cleaned.csv'
df.to_csv(output_path, index=False)

size_mb = os.path.getsize(output_path) / (1024 * 1024)
print(f"\n✅ Done! Saved: {output_path}")
print(f"   Size  : {size_mb:.1f} MB")
print(f"   Rows  : {len(df)}")
print(f"   Cols  : {df.shape[1]}")
print(f"\nColumns: {df.columns.tolist()}")
print("\nSample:")
print(df.head())
```
